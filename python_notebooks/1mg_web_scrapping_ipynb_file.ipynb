```python

 # Import necessary libraries

from bs4 import BeautifulSoup      # For parsing HTML content
import requests                    # For making HTTP requests
import pandas as pd                # For data manipulation and analysis
from time import sleep             # For introducing delays in web scraping
import time
```

```python

# Define the base URL for web scraping
base_url = 'https://www.1mg.com'

# Create an empty list to store product URLs
a = []

# Loop through a range of page numbers (in this case, just one page)
for i in range(1,2):

    # Construct the URL for the specific page
    url = f'{base_url}/categories/homeopathy-57?filter=true&pageNumber=={i}'

    # Define user-agent headers to mimic a web browser
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}

    sleep(0.05)

    # Send an HTTP GET request to the URL with specified headers
    req = requests.get(url, headers = headers)

     # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(req.content, 'html.parser')

    soup = BeautifulSoup(req.content)

    # Find all product boxes on the page
    products = soup.find_all('div',class_='style__product-box___liepi')

    # Loop through each product box
    for product in products:

        # Extract the URL of the product and append it to the list
        url = product.find('a',class_='style__product-link___UB_67').get('href')
        link = 'https://www.1mg.com'+url
        a.append(link)
    sleep(0.05)
```

```python

# Define a time limit for web scraping requests
time_limit = 1

# Initialize empty lists to store data
Name = []
Brand_name = []
Key_benefits = []
Key_ingredients = []
Rating = []
Number_of_rating = []

# Loop through each product URL in the list
for link in a:
    # Send an HTTP GET request to the product page
    req1 = requests.get(link, headers=headers)

    # Parse the HTML content of the product page using BeautifulSoup
    soup1 = BeautifulSoup(req1.text, 'html.parser')

    # Extract and store the product name
    name = soup1.find('h1',class_='ProductTitle__product-title___3QMYH')
    name=name.text.strip() if name else ""
    Name.append(name)

    # Extract and store the brand name
    Brand1 = soup1.find('div',class_='ProductTitle__manufacturer___sTfon')
    Brand = Brand1.text.strip() if Brand1 else ""
    Brand_name.append(Brand)

    # Extract and store the key benefits
    benefit = soup1.find('strong',string='Key Benefits:')
    benefits = benefit.find_next('li').string.strip() if benefit else ''
    Key_benefits.append(benefits)

    # Extract and store the key ingredients
    bag = soup1.find('strong', string = 'Key Ingredients:')
    key_ingredients = bag.find_next('li').string.strip() if bag else ''
    Key_ingredients.append(key_ingredients)

    # Extract and store the product rating
    ratings = soup1.find('div',class_='RatingDisplay__ratings-container___3oUuo')
    ratings=ratings.text.strip() if ratings else ""
    Rating.append(ratings)

    # Extract and store the number of ratings
    rating_number = soup1.find('span', class_='RatingDisplay__ratings-header___ZNj5b')
    rating_number=rating_number.text.strip() if rating_number else ""
    Number_of_rating.append(rating_number)

    # Introduce a small delay before processing the next product URL
    time.sleep(0.05)


# Create a dictionary to store the scraped data
data ={'Name':Name,'Brand_name':Brand_name,'Key_benefits':Key_benefits,'Key_ingredients':Key_ingredients, 'Rating':Rating,'Number_of_rating':Number_of_rating}

# Create a DataFrame from the scraped data
df = pd.DataFrame(data)
df
```
